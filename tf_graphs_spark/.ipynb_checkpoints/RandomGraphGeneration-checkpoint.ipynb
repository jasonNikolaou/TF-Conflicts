```python
import networkx as nx
import numpy as np
from collections import deque
import random
import pickle
from scipy.stats import truncnorm, truncexpon
import matplotlib.pyplot as plt
```

### Basic parameters

```python
n = 100 # num of individuals
k = 5 # num of individuals per team
numOfProjects = int(n / k) + 1
capacities = [k] * numOfProjects
mainProjectDist = "trunc_exponential" # uniform or trunc_exponential
maxSecondaryProjects = 3
secondaryProjectDist = "trunc_gaussian" # uniform or trunc_gaussian
conflictGraphType = "diversity" # friends or diversity
```

### Helper functions

```python
# helper function to sample main project for each team
def trunc_exponential(a=0, b=numOfProjects+1):
    # Generate N samples from a truncated exponential distribution
    # between a and b.
    # `b - a` specifies the range for the truncated distribution.
    # Set the `loc` parameter to 'a' and `scale` parameter to 1 for the exponential distribution.
    samples = truncexpon.rvs(b - a, loc=a, scale=10, size=1000)
    bins = np.arange(a, b)
    dist, bin_edges = np.histogram(samples, bins=bins, density=True)   
    return dist

def uniform(a=0, b=numOfProjects):
    return [1 / (b-a)] * (b-a)

def sample_main_project(dist="uniform"):
    if dist == "uniform":
        return uniform()
    elif dist == "trunc_exponential":
        return trunc_exponential()
```

```python
# helper functions to sample secondary projects for each member of a team
def trunc_gaussian(mu, lower=0, upper=numOfProjects+1, sigma=2):
    mu += 0.5 # center gaussian in the middle of the corresponding bin
    X = truncnorm(
        (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
    r = X.rvs(1000)
    bins = np.arange(lower, upper)
    dist, bin_edges = np.histogram(r, bins=bins, density=True)
    return dist

def sample_secondary_project(mu, dist="trunc_gaussian"):
    return trunc_gaussian(mu)
```

###  Teams and project preferences

```python
projects = {}
for i in range(numOfProjects):
    projects[i] = k
teams = deque()
for i in range(k, n+1, k):
    teams.append(list(range(i-k, i)))

# project preferences
c = dict()
for team in teams:
    # sample main project preference for team
    proj = np.random.choice(range(numOfProjects), p=sample_main_project(mainProjectDist))
    for u in team:
        c[(u, proj)] = 1
    
    # choose secondary projects (up to maxSecondaryProjects)
    dist = sample_secondary_project(proj, secondaryProjectDist)
    for u in team:
        ps = np.random.choice(dist, maxSecondaryProjects)
        for p in ps:
            c[(u, p)] = 1
        
```

###  Conflict graph

```python
w = dict() # w[(u, v)] = the weight of conflict edge w_{uv}
if conflictGraphType == "friends":
    p = 0.9 # probability of friendship between team members
    q = 1 / (n*2) # probability of friendship between non team members

    # friend graph
    G = nx.Graph()
    for u in range(n):
        for v in range(n):
            if u == v:
                continue
            if u // k == v // k: # u and v are in the same team
                rand = random.uniform(0, 1)
                if rand < p:
                    G.add_edge(u, v)
            else: # u and v are in different teams
                rand = random.uniform(0, 1)
                if rand < q:
                    G.add_edge(u, v)

    # conflict graph
    G_c = nx.complement(G)
    edges = list(G_c.edges)
    for e in edges:
        w[e] = 1

if conflictGraphType == "diversity":
    typesA = list(range(n))[::2]
    typesB = list(range(n))[1::2]
    print(typesA)
    print(typesB)

    # build conflict graph
    edgesA = [(u, v) for u in typesA for v in typesA if u < v]
    edgesB = [(u, v) for u in typesB for v in typesB if u < v]

    G = nx.Graph()
    G.add_edges_from(edgesA)
    G.add_edges_from(edgesB)
    edges = list(G.edges)
    w = dict()
    for e in edges:
        w[e] = 1
```

### Save generated instance

```python
with open(f'./inputs/graph_n{n}_{conflictGraphType}_{mainProjectDist}_{secondaryProjectDist}.pickle', 'wb') as file:
    pickle.dump(capacities, file)
    pickle.dump(c, file) # project preferences
    pickle.dump(w, file) # weights of conflict edges
    pickle.dump(edges, file) # conflict edges
    file.close()
```

### Visualize data

```python
# Plot project populatiry distribution
ys = sample_main_project(mainProjectDist)
xs = range(numOfProjects)


plt.bar(xs, ys)
plt.xlabel('project')
plt.ylabel('probability')
plt.title('Project popularity distribution')

plt.show()
```

```python
# plot project popularity for a single team
mainProject = 15
xs = range(numOfProjects)
ys = sample_secondary_project(mainProject)

plt.bar(xs, ys)
plt.xlabel('project')
plt.ylabel('probability')
plt.title(f'Secondary project selection (mainProject = {mainProject})')

plt.show()
```

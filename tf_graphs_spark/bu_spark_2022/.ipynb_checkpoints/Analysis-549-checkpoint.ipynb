## Input analysis

```python
def project_term(x):
    # Requires definition of lambda_
    res = 0
    # project preference term
    for u in nodes:
        for p in projects:
            if (u, p) in c:
                res += c[(u, p)] * x[(u, p)]
    res *= lambda_
    return res

def conflict_term(x):
    res = 0
    for e in edges:
        u, v = e
        inner_sum = 0
        for p in projects:
            inner_sum += x[(u, p)] * x[(v, p)]
        res += w[(u, v)] * (1 - inner_sum)
    return res
    
def f(x):
    # Requires definition of lambda_
    return project_term(x) + conflict_term(x) 
```

### Reconstruct optimal teams

```python
def construct_teams(x):
    teams = dict()
    for p in projects:
        teams[p] = []

    for e in x:
        u, p = e
        if x[e] == 1:
            u, p = e
            teams[p].append(u)
            
    for p in projects:
        teams[p].sort()
        
    return teams
            
teams_opt = dict()
for l in lambdasToString:    
    teams_opt[l] = construct_teams(x_opt[l])
    l_val = l.replace('p', '.')
    print(f'\n======== lambda {l_val} ==========')
    print(teams_opt[l])
```

### Compare $\lambda \sum_v \sum_t c_{vt}x_{vt}$ and $w(E)$
If $\alpha = \lambda \sum_v \sum_t c_{vt}x_{vt} - w(E) \geq 0$, then we can apply the $3/4$-approximation algorithm.
Otherwise, the guarantees of the algorithm do not hold.

```python
print(f'w(E) = {wE}')
for l, lstr in zip(lambdas, lambdasToString):
    lambda_ = l * wE / n
    projectsVal = project_term(x_round34[lstr])
    print(f'\n===== lambda = {l} ======')
    print(f'projects term value = {projectsVal}')
    print(f'a = {projectsVal - wE}')
```

### How many students got a project they like?

```python
def calc_preference_stats(teams):
    res = 0
    students = []
    for t in teams:
        team = teams[t]
        for u in team:
            res += 1 if (u, t) in c else 0
            students.append(u)
    res /= n
    return res, students


for l in lambdasToString:
    print(f'\n====== lambda = {l_val} =======')
    
    teams = teams_opt[l]
    res, students = calc_preference_stats(teams)
    
    l_val = l.replace('p', '.')
    print(f'(Optimal teams) percentage of students that got a project they like = {res}')
    print(f'list of students that got a project they like: {students}')
    
    teams = teams_round12[l]
    res, students = calc_preference_stats(teams)
    
    l_val = l.replace('p', '.')
    print(f'(1/2 approx alg) percentage of students that got a project they like = {res}')
    print(f'list of students that got a project they like: {students}')
    
    teams = teams_round34[l]
    res, students = calc_preference_stats(teams)
    
    l_val = l.replace('p', '.')
    print(f'(3/4 approx alg) percentage of students that got a project they like = {res}')
    print(f'list of students that got a project they like: {students}')
```

## Comparison with manual solution

```python
def min_friends(teams):
    minFriends = n
    for team in teams.values():
        for u in team:
            friends = sum([1 for v in team if u != v and (u, v) not in w and (v, u) not in w])
            minFriends = min(friends, minFriends)
    return minFriends
            
def max_friends(teams):
    maxFriends = 0
    for team in teams.values():
        for u in team:
            friends = sum([1 for v in team if u != v and (u, v) not in w and (v, u) not in w])
            maxFriends = max(friends, maxFriends)
    return maxFriends

def avg_friends(teams):
    friends = 0
    for team in teams.values():
        friends += sum([1 for u in team for v in team if u != v and (u, v) not in w and (v, u) not in w])
    return friends / n
        
def min_preference(teams):
    ans = 1
    for t in teams:
        team = teams[t]
        for u in team:
            ans = min(ans, 1 / c[(u, t)])      
    return ans
            
def max_preference(teams):
    ans = -1
    for t in teams:
        team = teams[t]
        for u in team:
            ans = max(ans, 1 / c[(u, t)])      
    return ans

def avg_preference(teams):
    pref = 0
    for t in teams:
        team = teams[t]
        pref += sum([1 / c[(u, t)] for u in team])
    return pref / n
```

```python
def group_keys_by_value(input_dict):
    grouped_dict = {}
    
    for key, value in input_dict.items():
        if value in grouped_dict:
            grouped_dict[value].append(key)
        else:
            grouped_dict[value] = [key]
    
    return grouped_dict
teams_manual = group_keys_by_value(assignment)
```

### Distribution of number of (in team) friends

```python
def friends_distribution(teams):
    dist = deque()
    for team in teams.values():
        for u in team:
            friends = sum([1 for v in team if u != v and (u, v) not in w and (v, u) not in w])
            dist.append(friends)
    return dist
```

```python
bins = [0, 1, 2, 3, 4, 5]
optDist = friends_distribution(teams_optimal)
manualDist = friends_distribution(teams_manual)
plt.hist(manualDist, bins=bins)
plt.xlabel("# friends")
plt.ylabel("# students")
plt.title("Number of (in-team) friends (manual)")
plt.show()
plt.hist(optDist, bins=bins)
plt.xlabel("# friends")
plt.ylabel("# students")
plt.title("Number of (in-team) friends (optimal)")
plt.show()
```

### Distribution of project preferences

```python
def project_distribution(teams):
    dist = deque()
    for t in teams:
        team = teams[t]
        for u in team:
            dist.append(1 / c[(u, t)])
    return dist
```

```python
bins = range(1, 11)
optDist = project_distribution(teams_optimal)
manualDist = project_distribution(teams_manual)
plt.hist(manualDist, bins=bins)
plt.xlabel("project preference")
plt.ylabel("# students")
plt.title("proejct preferences (manual)")
plt.show()
plt.hist(optDist, bins=bins)
plt.xlabel("project preference")
plt.ylabel("# students")
plt.title("project preferences (optimal)")
plt.show()
```

### Per student analysis

```python
def find_friends(student, teams):
    if student not in nodes:
        return "Not a valid student"
    friends = []
    for team in teams.values():
        for u in team:
            if u == student:
                friends = [v for v in team if u != v and (u, v) not in w and (v, u) not in w]
                return friends
    return []

def find_preference(student, teams):
    if student not in nodes:
        return "Not a valid student"
    for t in teams:
        team = teams[t]
        for u in team:
            if u == student:
                return 1 / c[(u, t)] 
    return -1
```

```python
equal = 0
opt_better = 0
manual_better = 0
tradeOff = 0

for u in nodes:
    print(f"=== student: {u} ===")
    opt_pref = find_preference(u, teams_optimal)
    manual_pref = find_preference(u, teams_manual)
    opt_friends = find_friends(u, teams_optimal)
    manual_friends = find_friends(u, teams_manual)
    
    
    print(f"optimal: friends = {opt_friends}, preference = {opt_pref}")
    print(f"manual: friends = {manual_friends}, preference = {manual_pref}")
    
    if opt_pref == manual_pref and len(opt_friends) == len(manual_friends):
        print("equal")
        equal += 1
    elif opt_pref <= manual_pref and len(opt_friends) >= len(manual_friends):
        print("opt better")
        opt_better += 1
    elif opt_pref >= manual_pref and len(opt_friends) <= len(manual_friends):
        print("manual better")
        manual_better += 1
    else:
        print("trade-off")
        tradeOff += 1

        
print("=== Stats ===")
print(f'equal = {equal}')
print(f'opt better = {opt_better}')
print(f'manual better = {manual_better}')
print(f'trade-off = {tradeOff}')
      
```

### Friends graph colored based on team assignment

```python
pos=nx.spring_layout(G_c, k=0.2, iterations=20)
def draw_graph_with_teams(teams, title):
    colors = sns.color_palette("tab10", len(teams))
    colors_dict = dict(zip(sorted(teams.keys()), colors))

    color_map = []
    for node in G:
        for team in teams:
            if node in teams[team]:
                color_map.append(colors_dict[team])
                break
                
    nx.draw(G_c, node_color=color_map, pos=pos, with_labels=True)
    for node in G:
        if representatives[node]:
            x, y = pos[node]
            plt.text(x + 0.09, y + 0.01, '*', fontsize=12, color='black', ha='center', va='center')
            
    plt.title(title)
    legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=label, markersize=10, markerfacecolor=color, markeredgecolor=color) for color, label in zip(colors, colors_dict.keys())]
    plt.legend(handles=legend_handles)
    plt.show()
```

```python
draw_graph_with_teams(teams_manual, "Friends graph and teams (manual assignemnt)")
draw_graph_with_teams(teams_optimal, "Friends graph and teams (optimal assignment)")
```
